```python
#Import
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read Data
df = pd.read_csv("../data/preprocessed/clean_data.csv")
```

```python
# Compute correlation matrix
corr = df.corr(numeric_only=True)

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))  # Adjust size as needed

# Draw the heatmap
sns.heatmap(
    corr,
    mask=mask,
    annot=True,
    fmt=".2f",          # Limit decimal places
    cmap="coolwarm",    # Better color contrast
    linewidths=0.5,     # Line between cells
    cbar_kws={"shrink": 0.8},  # Color bar size
    square=True
)

plt.title("Correlation Matrix", fontsize=16)
plt.xticks(rotation=45, ha="right")  # Rotate x-labels
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

StudyTimeWeekly to GPA

```python
# Define the bins and labels
bins = [0, 5, 10, 15, 20, 100]  # adjust max as needed
labels = ['1–5 hrs', '6–10 hrs', '11–15 hrs', '16–20 hrs', '21+ hrs']

# Create a new column for the study time range
df['StudyTimeRange'] = pd.cut(df['StudyTimeWeekly'], bins=bins, labels=labels, right=True)
```

```python
plt.figure(figsize=(10, 6))

# Bar plot of average GPA for each time range
sns.barplot(x='StudyTimeRange', y='GPA', data=df, estimator='mean', ci='sd')

plt.title('Average GPA by Study Time Range')
plt.xlabel('Study Time per Week')
plt.ylabel('Average GPA')
plt.tight_layout()
plt.show()
```

Absences to GPA

```python
# Define bins and labels for absences
absence_bins = [0, 2, 5, 10, 20, 100]  # You can adjust based on your dataset
absence_labels = ['0–2', '3–5', '6–10', '11–20', '21+']

# Create a new column for the absence range
df['AbsenceRange'] = pd.cut(df['Absences'], bins=absence_bins, labels=absence_labels, right=True)
```

```python
plt.figure(figsize=(10, 6))

# Bar plot of average GPA per absence group
sns.barplot(x='AbsenceRange', y='GPA', data=df, estimator='mean', ci='sd')

plt.title('Average GPA by Absence Range')
plt.xlabel('Absences')
plt.ylabel('Average GPA')
plt.tight_layout()
plt.show()
```
