```python
#Import
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read Data
df = pd.read_csv("../data/preprocessed/clean_data.csv")
```

Average GPA by Extracurricular Activity

```python
df.groupby('Extracurricular')['GPA'].mean()
```

```python
sns.barplot(x='Extracurricular', y='GPA', data=df)
```
